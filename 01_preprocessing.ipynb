```python
import pandas as pd
import numpy as np
import os

# Load dataset from data/raw
file_path = "data/raw/german_credit.csv"
df = pd.read_csv(file_path)

print("Data loaded successfully!")
print("Shape:", df.shape)
df.head()
```

```python
# Check null/missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Example: Fill numeric columns with median, categorical with mode
for col in df.columns:
    if df[col].dtype in ["int64", "float64"]:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled (no warnings).")
```

```python
# Save cleaned dataset to interim folder
os.makedirs("data/interim", exist_ok=True)
df.to_csv("data/interim/german_credit.csv", index=False)

print("Clean dataset saved to data/interim/german_credit.csv")
```
